REPLACING EMPTY VALUES

```python
dataset.replace('?',np.nan, inplace=True)
dataset.isnull().sum()
```

```python
dataset2 = dataset.copy()
```

HANDLE NULL DATA

```python
dataset2.fillna(dataset2.mean(numeric_only=True), inplace=True)
dataset2.isnull().sum()
```

```python
def assign_aqi_bucket(aqi):
    if pd.isna(aqi):  # If AQI itself is NaN, return NaN
        return None
    elif aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Satisfactory"
    elif aqi <= 200:
        return "Moderate"
    elif aqi <= 300:
        return "Poor"
    elif aqi <= 400:
        return "Very Poor"
    else:
        return "Severe"

dataset2["AQI_Bucket"] = dataset2.apply(lambda row: assign_aqi_bucket(row["AQI"]) if pd.isna(row["AQI_Bucket"]) else row["AQI_Bucket"], axis=1)
```
